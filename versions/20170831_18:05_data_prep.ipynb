```python
def create_pairset(all_samples, pairs):
    pairsets = pd.merge(all_samples[['entity:sample_id', 'tsca_id']], pairs, \
                    left_on='entity:sample_id', right_on='case_sample_id', \
                    how='inner')[['tsca_id', 'entity:pair_id']] \
                    .rename(columns={'tsca_id': 'membership:pair_set_id', 'entity:pair_id': 'pair_id'})
    return pairsets
```
